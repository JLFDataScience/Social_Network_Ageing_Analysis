### Procesado y limpieza de los datosüßπ

```python
df_TWenv = pd.read_csv('tweets_envejecimiento_full.csv')
```

```python
#Nos quedamos con las columnas que parecen m√°s interesantes
df_TWenv = df_TWenv[['tweet_created_at','tweet_id','user_screen_name','tweet_entities_hashtags',
                  'tweet_entities_mentions','tweet_full_text','tweet_retweet_count', 'user_followers_count','user_location']]
```

```python
#Observamos cuantos retweets hay con RT en el texto pero que no tenga valor en el campo 'tweet_entities_mentions'
df_TWenv[df_TWenv['tweet_full_text'].str.contains('RT ')]['tweet_entities_mentions'].isnull().sum()
```

```python
#Se comprueba que los RT pertenecen a hashtags o partes del texto, por tanto todos los que tienen RT, 
#tienen tweet_entities_mentions
rt = df_TWenv[df_TWenv['tweet_full_text'].str.contains('RT ')]
rt[rt['tweet_entities_mentions'].isnull()]
```

```python
#Los usuarios m√°s mencionados y retweeteados
df_TWenv[df_TWenv['tweet_full_text'].str.contains('RT')]['tweet_entities_mentions'].value_counts()
```

```python
#Contamos cuantos se mencionan frente a los RT
print('Los mencionados con valores nulos:',df_TWenv['tweet_entities_mentions'].isnull().sum())
print('Los RT sin valor en mencionados son:', df_TWenv[df_TWenv['tweet_full_text'].str.contains('RT')]['tweet_entities_mentions'].isnull().sum())
```

```python
#A√±adimos una columna para contabilizar si es RT (ReTweet)
pattern = re.compile(r'RT ')
df_TWenv['rt'] = df_TWenv['tweet_full_text'].apply(lambda x: True if pattern.findall(x) else False)
```

```python
#Generamos el df para el grafo quedando tres campos usuario (meci√≥n), mencionado
df_grafo =df_TWenv[['user_screen_name', 'tweet_entities_mentions', 'rt']]
```

```python
#Renombramos columnas
df_grafo.columns = ['emisor','mencionado','rt']
df_grafo.head()
```

```python
#Eliminamos las @
df_grafo['mencionado'] = df_grafo['mencionado'].str.replace(r'@', '')
df_grafo.head()
```

```python
#Comprobamos si hay m√°s de un mencionado por emisor en la columna 'mencionado'
len(df_grafo[df_grafo['mencionado'].str.contains(',', na=False)])
```

```python
#Comprobamos que tipo de estructura tienen esas m√∫ltiples menciones
df_grafo[df_grafo['mencionado'].str.contains(',', na=False)][0:10]
```

```python
#Genero un nuevo df a√±adiendo filas nuevas cuando hay m√°s de un mencionado
df_grafo = df_grafo.assign(mencionado=df_grafo['mencionado'].str.split(',')).explode('mencionado').reset_index(drop=True)
#Compruebo que no hay m√°s de un nombre en la columna 'mencionado'
len(df_grafo[df_grafo['mencionado'].str.contains(',', na=False)])
```

```python
df_grafo[:10]
```

```python
#Al haber dividido por la ',', quedar√°n espacios en el inicio del nombre. Intentamos eliminarnos para homogeneizar nombres
df_grafo['emisor'] = df_grafo['emisor'].str.replace(r' ', '')
#Eliminamos tambi√©n los gui√≥n bajo
df_grafo['mencionado'] = df_grafo['mencionado'].str.replace(r'_', ' ')
df_grafo['emisor'] = df_grafo['emisor'].str.replace(r'_', ' ')
df_grafo[:10]
```

### Se genera el Grafo

```python
import networkx as nx
#Formamos el grafo con el dataframe: emisor, mencionado. Si hay RT-> mismo comportamiento que si no lo hay
G = nx.DiGraph()

for i in df_grafo.index.values:
     G.add_edge(df_grafo.loc[i]['emisor'],df_grafo.loc[i]['mencionado'])
    
nx.write_graphml(G, "Tweet_envejecimiento.graphml")
```

```python
print("N¬∫ nodos",len(G.nodes.data()))
print("N¬∫ enlaces",len(G.edges.data()))
```

### M√©tricas del gr√°fo

#### ¬øEs conexo?

```python
#Transformamos el grafo a No dirigido para realizar esta comprobaci√≥n
GU = G.to_undirected()
print(nx.is_connected(GU))
```

#### Densidad de la red

```python
#A priori, la densidad de la red deber√≠a ser muy baja ya que no tenemos enlaces de seguimiento, 
#por lo que muchos nodos s√≥lo emitir√°n pero no recibir√°n enlace
nx.density(G)
```

#### Top 5 m√°s mencionados

```python
#Calculo nodos con m√°s aristas de entrada
in_degree=list(G.in_degree())
in_degree.sort(key=lambda x: (x[1]),reverse=True)
in_degree[1:5]
```

#### Top 5 eigenvector

```python
#Esta medida tiene en cuenta c√≥mo de cerca est√° un nodo del resto de nodos importantes de la red
#lo que le puede hacer m√°s influyente
e=nx.eigenvector_centrality(G)
e=list(e.items())
e.sort(key=lambda x: (x[1]),reverse=True)
e[0:5]
```

#### M√°x/m√≠n y media grado de la red

```python
degree_list = []
for v in G:
    degree_list.append(G.degree(v))

print('Max degree:')
print(max(degree_list))
print('Min degree:')
print(min(degree_list))
print('Average degree:')
print(sum(degree_list)/len(degree_list))
```

#### Centralidad de Clustering

```python
nx.average_clustering(G)
```

```python
nx.hits(G)[0:1]
```

#### Censo de triadas

```python
nx.triadic_census(G)
#La mayor√≠a son triadas desconectadas (003) y la segunda son triadas que conectan s√≥lo dos miembros
```
